```python
# Select features and target variable

target = "salary_usd"

features = [
    "ai_specialization",
    "experience_level",
    "experience_years",
    "industry",
    "country",
    "company_size",
    "education_required",
    "work_mode",
    "ai_adoption_score",
    "company_funding_billion",
    "economic_index",
    "ai_maturity_years",
    "skill_demand_score",
    "cost_of_living_index",
    "company_rating",
    "career_growth_score",
    "job_security_score",
    "promotion_speed"
]

X = df[features]
y = df[target]
```

```python
# identify categorical and numerical features

categorical_cols = X.select_dtypes(include="object").columns

numeric_cols = X.select_dtypes(include=["int64","float64"]).columns

print("Categorical:", categorical_cols)
print("Numeric:", numeric_cols)
```

```python
# Encode categorical features, will use one-hot encoding for simplicity

from sklearn.preprocessing import OneHotEncoder

X_encoded = pd.get_dummies(X, drop_first=True)
print(X_encoded.head())
```

```python
# train-test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
# Baseline model - Multiple Linear Regression

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Evaluate the model

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

r2 = r2_score(y_test,y_pred)

mae = mean_absolute_error(y_test,y_pred)

rmse = np.sqrt(mean_squared_error(y_test,y_pred))

print("R2:",r2)
print("MAE:",mae)
print("RMSE:",rmse)
```

```python
# Residual analysis

residuals = y_test - y_pred

plt.scatter(y_pred,residuals)

plt.axhline(0)

plt.xlabel("Predicted Salary")

plt.ylabel("Residual")

plt.show()
```

```python
What you are describing is a classic pattern called Heteroscedasticity. In a residual plot, if the dots start close together and fan out (like a funnel or a "v" shape) as the predicted salary increases, it means your model's error is not constant.
Technically, it violates one of the core assumptions of Linear Regression (the assumption of Homoscedasticity).
What this means for your data:
It doesn't necessarily mean the data "isn't meant" for regression, but it means a Simple Linear Regression is the wrong tool for the job. It suggests that as experience/salary grows, the "noise" or variation grows too. (e.g., Entry-level salaries are all very similar, but Senior-level salaries vary wildly).
```

```python
# We log the Target (y) to compress the "fanning" effect
df['Log_Salary'] = np.log(df['salary_usd'])

# 3. Prepare Models
y_raw = df['salary_usd']
y_log = df['Log_Salary']

# Split Data
X_train, X_test, y_train_raw, y_test_raw = train_test_split(X_encoded, y_raw, test_size=0.2)
X_train, X_test, y_train_log, y_test_log = train_test_split(X_encoded, y_log, test_size=0.2, random_state=42)

# Fit Models
model_raw = LinearRegression().fit(X_train, y_train_raw)
model_log = LinearRegression().fit(X_train, y_train_log)

# 4. Visualization: Comparing Residuals
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Raw Data Residuals (The "Funnel" you saw)
preds_raw = model_raw.predict(X_test)
residuals_raw = y_test_raw - preds_raw
ax1.scatter(preds_raw, residuals_raw, color='red', alpha=0.6)
ax1.axhline(0, color='black', linestyle='--')
ax1.set_title('Original Data: Heteroscedasticity (Funnel)')
ax1.set_xlabel('Predicted Salary')
ax1.set_ylabel('Residuals')

# Plot 2: Log-Transformed Residuals (The Fix)
preds_log = model_log.predict(X_test)
residuals_log = y_test_log - preds_log
ax2.scatter(preds_log, residuals_log, color='green', alpha=0.6)
ax2.axhline(0, color='black', linestyle='--')
ax2.set_title('Log-Transformed: Homoscedasticity (Even Cloud)')
ax2.set_xlabel('Predicted Log Salary')
ax2.set_ylabel('Residuals')

plt.tight_layout()
plt.show()

# 5. How to convert back to actual dollars for your final prediction
# If you predict a value, use np.exp() to reverse the log
sample_pred_log = model_log.predict(X_test.iloc[[0]])  # Predicting for the first test sample
actual_dollar_pred = np.exp(sample_pred_log)
print(f"Predicted Salary for {X_test.iloc[0][['experience_years', 'skill_demand_score']]}: ${actual_dollar_pred[0]:,.2f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures

# 1. Create Polynomial Features (degree=2 adds an 'Experience squared' column)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_encoded) # X is your Experience column

# 2. Fit the model using these new features and your Log_Salary
model_poly = LinearRegression().fit(X_poly, y_log)

# 3. Check residuals again
preds_poly = model_poly.predict(X_poly)
residuals_poly = y_log - preds_poly

# Plotting this should give you the "random snowstorm" cloud you want
plt.scatter(preds_poly, residuals_poly, color='blue', alpha=0.5)
plt.axhline(0, color='black', linestyle='--')
plt.title('Polynomial + Log: Final Clean Residuals')
plt.show()
```
