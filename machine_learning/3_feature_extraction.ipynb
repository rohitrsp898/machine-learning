```python
import pandas as pd

data = {
    "customer_id": [101,102,103,104,105],
    "purchase_date": [
        "2024-01-05",
        "2024-02-10",
        "2024-03-15",
        "2024-03-20",
        "2024-04-02"
    ],
    "age": [25, 40, 35, 28, 50],
    "gender": ["Male", "Female", "Female", "Male", "Female"],
    "purchase_amount": [500, 1200, 750, 300, 2000],
    "is_subscribed": [1,0,1,0,1]
}

df = pd.DataFrame(data)

df["purchase_date"] = pd.to_datetime(df["purchase_date"])

print(df)
```

```python
# Feature Extraction

# Extracting month, day, and year from purchase_date
df["purchase_month"] = df["purchase_date"].dt.month
df["purchase_day"] = df["purchase_date"].dt.day
df["purchase_year"] = df["purchase_date"].dt.year

# Creating age groups
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 45:
        return "Middle"
    else:
        return "Senior"

# Applying the age_group function to create a new column
df["age_group"] = df["age"].apply(age_group)

# Encoding categorical variables
df["gender_encoded"] = df["gender"].map({
    "Male":0,
    "Female":1
})

# Creating a new feature for high value customers
df["high_value_customer"] = df["purchase_amount"].apply(
    lambda x: 1 if x > 1000 else 0
)

df
```

```python
# Feture Scaling

# Min-Max Scaling

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_columns = ["age","purchase_amount","purchase_month"]

df[scaled_columns] = scaler.fit_transform(df[scaled_columns])

print(df)
```

```python
# standard Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[["age_scaled","amount_scaled"]] = scaler.fit_transform(
    df[["age","purchase_amount"]]
)

df
```
